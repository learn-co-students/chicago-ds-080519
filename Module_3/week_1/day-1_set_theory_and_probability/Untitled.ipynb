```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
rand_nums = np.random.randint(1,7,100)
range_list = list(rand_nums)
freq = {}
for num in range_list:
    if num in freq:
        freq[num] =+1
    freq[num] = num
freq
```

```python
list_mean = []

for y in range(1,1001):
    number_list = []
    for x in range(101):
        number_list.append(np.random.randint(1,100))
    num_mean = np.mean(number_list)
    list_mean.append(num_mean)
```

```python
import seaborn as sns
sns.distplot(list_mean,kde=True)
plt.show()
```

```python
new_list = []
running_total = 0
for i in list_mean:
    running_total += i 
    new_list.append(running_total)
```

```python
cdf_list = [np.mean(number) for number in new_list]
plt.scatter(cdf_list, new_list)
```
