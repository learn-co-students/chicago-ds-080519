````python
cnx = sqlite3.connect('data/pokemon.db')
````

````python
# q1: query all columns from Pokemon the Pokemon that have base_experience above 200
q1 = '''
SELECT *
FROM pokemon
WHERE base_experience > 200
'''
pd.read_sql(q1, cnx)
````

````python
# q2: query the id, name, type1 and type2 of Pokemon that have water types as either their first or second type
q2 = '''
SELECT id, name, type1, type2
FROM pokemon
WHERE type1 = "water" OR type2 = "water"
'''
pd.read_sql(q2, cnx)
````

````python
# q3: query the average weight of Pokemon by their first type in descending order
q3 = '''
SELECT type1, AVG(weight) AS avg_weight
FROM pokemon
GROUP BY type1
ORDER BY avg_weight DESC
'''
pd.read_sql(q3, cnx)
````

````python
# q4: query the Pokemon name, Pokemon type2, and what type2 has "2xdamage" to
q4 = '''
SELECT pokemon.name, type2, "2xdamage"
FROM pokemon
LEFT JOIN types ON types.name=pokemon.type2
'''
df = pd.read_sql(q4, cnx)
print(df['2xdamage'].value_counts())
df
````

````python
# a test
df[df['type2'].isnull()]['2xdamage'].value_counts()
````

````python
# q5: query the top 5 most common type1s, the minimum height, maximum height, minimum weight and maximum weight of pokemon with those type1s, and what associated type they do "0.5xdamage" to
q5 = '''
SELECT type1, COUNT(type1) AS type_count, MIN(height), MAX(height), MIN(weight), MAX(weight), "0.5xdamage"
FROM pokemon
INNER JOIN types ON pokemon.type1=types.name
GROUP BY type1
ORDER BY type_count DESC
LIMIT 5
'''
pd.read_sql(q5, cnx)
````

2. Write a function to get **all** the authors and href links for the authors from the [homepage](http://quotes.toscrape.com/)


````python
def authors(url):
    '''
    input: url
    
    return: a dictionary of of authors and their urls
            {'author_1':'url_of_author_1', 'author_2':'url_of_author_2' ...}
    '''
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser')
    
    authors = soup.find_all(class_ = 'author')
    author_hrefs = soup.select('span a')
    
    authors = [a.get_text() for a in authors]
    author_hrefs = [url+link.get('href') for link in author_hrefs]
    
    author_dict = dict(zip(authors, author_hrefs))
    
    return author_dict
    
print(authors('http://quotes.toscrape.com'))
````

````python
# run this cell to test the function
print(authors('http://quotes.toscrape.com/'))
print('\n')
print(authors('http://quotes.toscrape.com/page/3'))
````

````python
# Your code here
first_authors = []
url_stem = 'http://quotes.toscrape.com'
url = url_stem
for i in range(5):
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser')
    first_authors.append(soup.find(class_='author').get_text())
    next_page_element = soup.find(class_='next')
    next_url = next_page_element.find('a').get('href')
    url = url_stem+next_url

print(first_authors)
````

4. Write a function to get all of the quotes from a page.

````python
def get_some_quotes(url, npages=1):
    '''
    input: url, number of pages to scrap (just scrape the home page if no argument is passed in)
    
    return: a list of dictionaries of quotes with their attributes
            [{'quote':'quote_1_text', 'author':'url_of_author_1'}, 
            {'quote':'quote_2_text', 'author':'url_of_author_2', 'quote_tags':[list_of_quote_2_tags]}, ...]
    '''
    url_stem = url
    page_link = url_stem
    quote_dict_list = []
    for i in range(npages):
        html = requests.get(page_link)
        soup = BeautifulSoup(html.content, 'html.parser')
        quotes = soup.find_all(class_='quote')
        
        for q in quotes:
            quote_text = q.find('span', class_='text').get_text()            
            author = q.find(class_='author').get_text()          
            tags = [t.get_text() for t in q.find_all('a', class_='tag')]
            
            quote_dict = {'quote': quote_text, 'author': author, 'quote_tags': tags}
            quote_dict_list.append(quote_dict)
        
        
        next_page_element = soup.find(class_='next')
        next_url = next_page_element.find('a').get('href')
        page_link = url_stem + next_url
    
    return quote_dict_list
        
        
get_some_quotes('http://quotes.toscrape.com', 2)
    
````

````python
# set the function to a variable to use later
quotes_for_mongo = get_some_quotes('http://quotes.toscrape.com')
quotes_for_mongo
````
