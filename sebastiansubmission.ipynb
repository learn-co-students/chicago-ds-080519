````python
def get_pokedata(url):
    
    """
    url is a string of the URL that will get you the relevant info from the API
    for a single pokemon. 
    Example: "https://pokeapi.co/api/v2/pokemon/1/"
    
    this function should return the dictionary for ONE pokemon
    
    """
    pokeurl = url
    results = requests.get(pokeurl).json()
    poke_id = results['id']
    poke_name = results['name']
    poke_base_experience = results['base_experience']
    poke_weight = results['weight']
    poke_height = results['height'] 
    poke_types_one = results['types'][0]['type']['name']
    if len(results['types']) == 2: 
        poke_types_two = results['types'][1]['type']['name']
    else:
        poke_types_two = np.nan
    
    to_df = [poke_id, poke_name.title(), poke_base_experience, poke_weight, poke_height, poke_types_one, poke_types_two]
    df = pd.DataFrame(to_df).T
    df.columns = ['id', 'name', 'base_experience', 'weight', 'height', 'type_one', 'type_two']
    return df
````

2. Write a function to get **all** the authors and href links for the authors from the [homepage](http://quotes.toscrape.com/)


````python
def authors(url):
    '''
    input: url
    
    return: a dictionary of of authors and their urls
            {'author_1':'url_of_author_1', 'author_2':'url_of_author_2' ...}
    '''
    
    html_page = requests.get(url)
    soup = BeautifulSoup(html_page.content, 'html.parser')
    names = [x.text for x in soup.findAll('small', class_ = 'author')]
    links = [x.next_sibling.next_sibling['href'] for x in soup.findAll('small', class_ = 'author')]
    author_dict = {}
    for i, name in enumerate(names):
        author_dict[name] = links[i]
    return author_dict
````

4. Write a function to get all of the quotes from a page.

````python
def get_some_quotes(orig_url, num_of_pages = 1):
    '''
    input: url, number of pages to scrape (just scrape the home page if no argument is passed in)
    
    return: a list of dictionaries of quotes with their attributes
            [{'quote':'quote_1_text', 'author':'url_of_author_1'}, 
            {'quote':'quote_2_text', 'author':'url_of_author_2', 'quote_tags':[list_of_quote_2_tags]}, ...]
    '''
    list_of_dicts = []
    for n in range(num_of_pages):
        url = orig_url + f"page/{n+1}/"
        html_page = requests.get(url)
        soup = BeautifulSoup(html_page.content, 'html.parser')
        
        page_list = []
        names = [x.text for x in soup.findAll('small', class_ = 'author')]
        links = [x.next_sibling.next_sibling['href'] for x in soup.findAll('small', class_ = 'author')]
        quotes = [x.text for x in soup.findAll('span', class_ = 'text')]   
        tags_list = [x['content'] for x in soup.findAll('meta', class_ = 'keywords')]
        for i in range(len(quotes)):
            entry_dict = {}
            quote = quotes[i]
            quote_num = f"page {n + 1} quote {i + 1}"
            name = names[i]
            if "." in name:
                name = name.replace(".", "")
            link = links[i]
            tags = tags_list[i]
            tags = tags.split(',')
            entry_dict[quote_num] = quote
            entry_dict[name] = link
            entry_dict['tags'] = tags
            page_list.append(entry_dict)
        for item in page_list:
            list_of_dicts.append(item)
    return list_of_dicts
````

````python
# set the function to a variable to use later
quotes_for_mongo = get_some_quotes('http://quotes.toscrape.com/', 5)
quotes_for_mongo
````
